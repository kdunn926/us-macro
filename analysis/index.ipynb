## Overview

The purpose of this project is to aggregate and disseminate publically available US macroeconomic data. Extensive use of data APIs and web-scraping 

All the results and plots presented in the pages below aim to be reproducible and reference the primary-source data, directly whenever possible. Follow the
[Setup Instructions](setup.html) if you are interested in reproducing the results for yourself.

The following list contains automatically-updated interactive plots for each category. All content is generated by rendering the [Jupyter notebooks](https://github.com/kdunn926/us-macro/tree/master/analysis) into webpages.

- ### [Consumer Spending Charts](analysis/consumer.html) - aka "70%" of the US economy
   Source: [US Census Bureau / Monthly Retail Trade Report](https://www.census.gov/retail/index.html)

- ### [Manufacturing Charts](analysis/manufacturing.html)
   Sources:
     - [CSX Freight Reports](https://www.csx.com/index.cfm/investors/volume-trends/),
     - [CASS Transportation Indexes](https://www.cassinfo.com/freight-audit-payment/cass-transportation-indexes)
<br><br>
- ### [Labor Market Charts](analysis/nonfarmpayrolls.html)
   Source: [US Bureau of Labor Statistics / Current Employment Statistics](https://www.bls.gov/ces/)

- ### [Residential Real Estate Charts](analysis/residential-homes.html)
   Source: [Case/Shiller Home Price Index](http://www.econ.yale.edu/~shiller/data/Fig2-1.xls)
   
- ### [Commercial Credit Charts](analysis/commercial-credit.html)
   Source: [Federal Reserve Bank of St. Louis - FRED](https://fred.stlouisfed.org/graph/?g=pZ8g)
